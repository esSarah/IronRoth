```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Data/Data_Marketing_Customer_Analysis_Round3.csv')
```

```python
df.describe()
```

```python
numerical = df.select_dtypes(np.number)
```

```python
cathegorical = df.select_dtypes(object)
```

```python
numerical
```

```python
numerical.columns
```

```python
cols=['customer_lifetime_value', 'income',
       'monthly_premium_auto', 'months_since_last_claim',
       'months_since_policy_inception', 'number_of_open_complaints',
       'number_of_policies', 'total_claim_amount']
```

```python
sns.pairplot(numerical)
```

```python
for col in cols:
    sns.displot(numerical, x=col, bins=20)
```

```python
counter = 0
plt.figure(figsize=(20,20))
# fig, ax = plt.subplots(5,5)
for col in numerical.columns:
    counter +=1
    plt.subplot(5,5,counter)
    plt.title(col)
    plt.hist(df[col], bins=20)
```

```python
# create figure and axes
fig2, ax2 = plt.subplots(figsize = (6,6))

# set title
ax2.set_title('Total Responce rate', fontweight='bold')

# plot the bar chart
ax2.bar(cathegorical['response'].value_counts().index, cathegorical['response'].value_counts(), color=('red','green'))
```

```python
cathegorical
```

```python
response_by_sales = cathegorical.loc[:,['sales_channel', 'response']].groupby(['sales_channel'])['response'].value_counts().reset_index(name ='cahnnel_responce')
```

```python
sns.set_palette('RdYlGn')
palette_alive = {"yes": "C5", "no": "C0"}
```

```python
fig3, ax3 = plt.subplots(figsize = (12, 8))
ax3.set_title('Response rate by the sales channel', size='25', fontweight='bold')
sns.barplot(x = response_by_sales['sales_channel'], y = response_by_sales['cahnnel_responce'], hue = response_by_sales['response'], ax = ax3, palette=palette_alive)
```

```python
sns.countplot(x = 'sales_channel', hue = 'response', data = cathegorical, palette=palette_alive )
```

### Show a plot of the response rate by the total claim amount.

```python
fig4, ax4 = plt.subplots(figsize = (15, 9))
sns.countplot(x = pd.cut(numerical['total_claim_amount'], bins= 8), hue = cathegorical.response, ax = ax4, palette=palette_alive) 
```

### Show a plot of the response rate by income.

```python
fig5, ax5 = plt.subplots(figsize = (15, 8))
sns.countplot(x = pd.cut(numerical['income'], bins = 8), hue = cathegorical.response, ax = ax5, palette = palette_alive)
```

```python
def correlation_barplot(col1, col2):
#     fign, axn = plt.subplots(figsize = (15, 8))
    sns.countplot(x = col1, hue = col2)
```

```python
correlation_barplot(pd.cut(numerical['income'],bins =8) , cathegorical['coverage'])
```

```python
correlation_barplot(cathegorical['education'], cathegorical['coverage'])
```

```python
correlation_barplot(pd.cut(numerical.income, bins = 8), cathegorical.gender)
```

## Correlation Heatmap.

```python
df.corr()
```

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style('white'):
    f, ax = plt.subplots(figsize=(9, 7))
    ax = sns.heatmap(corr, mask=mask, cmap='coolwarm', vmin= -1, vmax=1, annot=True, square=True)
```
